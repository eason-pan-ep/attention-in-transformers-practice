### Import Modules

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

### Self-Attention Class

```python
class SelfAttention(nn.Module):
    # nn.Module is the base class for all neural network modules you made with PyTorch

    def __init__(self, d_model=2, row_dim=0, col_dim=1):
        """
            Class constructor
            Params:
                d_model: the dimension of the model / word embedding values per token
                row_dim: row index in the data
                col_dim: column index in the data
        """

        super().__init__()

        # Define the learnable weight matrices for Q, K, V (untrained)
        self.W_q = nn.Linear(in_features=d_model, out_features=d_model, bias=False)
        self.W_k = nn.Linear(in_features=d_model, out_features=d_model, bias=False)
        self.W_v = nn.Linear(in_features=d_model, out_features=d_model, bias=False)
        # in_features defines how many rows are in Weight matrix
        # out_features defines how many columns in the Weight matrix
        # Following the "Attention is All You Need" not implementing bias terms

        self.row_dim = row_dim
        self.col_dim = col_dim

    def forward(self, token_encodings):
        """
            Forward pass of the self-attention layer
            Params:
                token_encodings: the input token encodings to the self-attention layer
            Returns:
                attention_scores: the output attention scores from the self-attention layer
        """
        q = self.W_q.forward(token_encodings)
        k = self.W_k.forward(token_encodings)
        v = self.W_v.forward(token_encodings)

        similarities = torch.matmul(q, k.transpose(self.row_dim, self.col_dim))
        scaled_sims = similarities / torch.sqrt(torch.tensor(k.size(self.col_dim)))
        attention_percentages = F.softmax(scaled_sims, dim=self.col_dim)
        attention_scores = torch.matmul(attention_percentages, v)

        return attention_scores
    
```

### Calculate Self-Attention

```python
encodings_mat = torch.tensor(
    [[1.16, 0.23],
     [0.57, 1.36],
     [4.41, -2.16]])

torch.manual_seed(42) # For reproducibility, set the seed for random number generator

self_attention_layer = SelfAttention(d_model=2, row_dim=0, col_dim=1)
self_attention_score = self_attention_layer.forward(encodings_mat)
print(self_attention_score)

## the last bit of the printed tensor is used for training the weights with backpropagation
```
